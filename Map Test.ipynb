## Creating a data driven map using lat and lon of high consumer off-gas postcodes

The CSV file has been created by merging sub-national electricity consumption data; off-gas grid postcodes; and geographical infomation data

```python
import pandas as pd
import numpy as np
from folium.plugins import MarkerCluster
MarkerCluster()
```

```python
df = pd.read_csv('GeoCdf2.csv')
df
```

```python
df_clean = df.groupby("Local Authority Name").first().reset_index()
print(df_clean)
```

```python
to_drop = ['Off-gas Consumption (kWh)',  'Year', 'Location']
df_clean.drop(to_drop, inplace=True, axis=1)
df_clean
```

```python
# to create a map of all local authorities with off-grid high electricity consumers (>Average consumption (kWh))
my_map = folium.Map(location=[51.753791, -0.330183], tiles='CartoDB positron', zoom_start=6)

marker_cluster = MarkerCluster().add_to(my_map)

for index, row in df_clean.iterrows():
    folium.Marker( 
        location=[row['Latitude'], row['Longitude']],
        popup=row['Local Authority Name'],
        tooltip=row['Local Authority Name']
    ).add_to(my_map)
    
my_map
```
