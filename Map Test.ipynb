## Creating a data driven map using lat and lon of high consumer off-gas postcodes

The CSV file has been created by merging sub-national electricity consumption data; off-gas grid postcodes; and geographical infomation data

```python
import pandas as pd
```

```python
!pip install folium 
import folium
```

```python
df = pd.read_csv('GeoCdf2.csv')
df
```

```python
my_map = folium.Map(location=[51.753791, -0.330183], zoom_start=6)

for _, POSTCODE in df.iterrows():
    folium.Marker( 
        location=[df['Latitude'], df['Longitude']],
        popup=df['Local Authority Name'],
        tooltip=df['Local Authority Name']
    ).add_to(my_map)

my_map
```
